```python
import numpy as np
signal = data.iloc[0, 1:-1].values
power = np.mean(signal**2)

print("Signal power:", power)
```

```python
normal_powers = []
seizure_powers = []

for i in range(len(data)):
    
    signal = data.iloc[i, 1:-1].values
    label = data.iloc[i]["y"]
    
    power = np.mean(signal**2)
    
    if label == 1:
        normal_powers.append(power)
        
    elif label == 5:
        seizure_powers.append(power)
```

```python
print("Average normal power:", np.mean(normal_powers))
print("Average seizure power:", np.mean(seizure_powers))
```

```python
plt.hist(normal_powers, alpha=0.5, label="Normal")
plt.hist(seizure_powers, alpha=0.5, label="Seizure")

plt.legend()
plt.title("EEG signal power comparison")
plt.xlabel("Power")
plt.ylabel("Count")

plt.show()
```
